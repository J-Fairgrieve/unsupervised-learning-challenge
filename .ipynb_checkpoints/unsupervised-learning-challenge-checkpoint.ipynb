```python
# Import the relevant libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

## 1. Data Preparation

```python
# Read the data into Pandas

df = pd.read_csv("crypto_data.csv", index_col=False)
df
```

```python
# Discard all cryptocurrencies that are not being traded, then remove the "IsTrading" column

clean_df = df[(df["IsTrading"] == True)]
clean_df = clean_df.drop(["IsTrading"], axis = 1)
```

```python
# Remove any rows that have at least one null value

clean_df = clean_df.dropna()
```

```python
# Filter to only show currency that has been mined

clean_df = clean_df[clean_df["TotalCoinsMined"] > 0]
```

```python
#  Remove the "CoinName" & "Unnamed: 0" columns

clean_df = clean_df.drop(["Unnamed: 0", "CoinName"], axis = 1)
clean_df
```

```python
# Convert the remaining columns with text values into numerical data

dummy_df = pd.get_dummies(clean_df, columns = ["Algorithm", "ProofType"], drop_first = True, dtype = float)
dummy_df
```

```python
# Standardise the dataset

scaler = StandardScaler()
scaled_data = scaler.fit_transform(dummy_df)
scaled_data
```
